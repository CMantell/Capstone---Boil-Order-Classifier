```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Import colormap from matplotlib
import matplotlib.cm

# Import mapping libraries
from mpl_toolkits.basemap import Basemap
from matplotlib.patches import Polygon # Patch for making polygons in map
from matplotlib.collections import PatchCollection # PatchCollection class for improving plotting speed
from matplotlib.colors import Normalize # Access the full range of the color map

#Plots figures in jupyter notebook
%matplotlib inline
```

```python
# Load dataframe - All waterborne illness in 2017 and marked with LDA infered topic
df = pd.read_csv('./GIS_example.csv')
df.head()
```

```python
# Make basemap
fig, ax = plt.subplots(figsize=(20, 20)) # Make matplotlib empty figure
lats = df.Lat.values  # Set latitude values
lons = df.Long.values # Set longitude values

# Basemap - resolution = crude, projection = Albers Equal Area, (Lat_0, lon_0) = center point of figure.  The rest are points for the corners of the figure
m=Basemap(resolution='c', projection='aea', lat_0=35, lon_0=-80, llcrnrlon=-115, llcrnrlat=18, urcrnrlon=-63, urcrnrlat=55)
x, y = m(lons, lats)
m.drawmapboundary(fill_color='#99ffff')
m.fillcontinents(color='#cc9966', lake_color='#99ffff')

# Plot points on map.  'bo' = 'Blue O's (circles)
m.plot(x,y, 'bo', markersize=4)
```

**Plots are successful but a little small.   Let's see how we can make it better...**

```python
# Make figure a little smaller
fig, ax = plt.subplots(figsize=(15, 15))

# Change resolution to full
m=Basemap(resolution='f', projection='aea', lat_0=35, lon_0=-80, llcrnrlon=-115, llcrnrlat=18, urcrnrlon=-63, urcrnrlat=55)

# Add states from shapefile
states = m.readshapefile('./tl_2010_us_state10', 'states')
x, y = m(lons, lats)
m.drawmapboundary(fill_color='#99ffff')
m.fillcontinents(color='#cc9966', lake_color='#99ffff')

# Make markers bigger
m.plot(x,y, 'bo', markersize=8)
```

**This looks nice but it's plots of one type.  Lets change the map so that all types of Waterborne Illnesses are plotted**

```python
# Let's look again at the different topics
df.topic.unique()
```

```python
# Let's seperate the different points by topic and make each a dataframe
df_hard = df.loc[(df.topic == 'Hard to classify')]
df_new = df.loc[(df.topic == 'New boil order issued or repaired')]
df_water = df.loc[(df.topic == 'Drinking water contaminated')]
df_exist = df.loc[(df.topic == 'Existing boil order continuing or lifted')]
df_beach = df.loc[(df.topic == 'Beach/Lake/Park closed from bacteria in water')]
```

```python
# Same map as before
fig, ax = plt.subplots(figsize=(15, 15))
m=Basemap(resolution='f', projection='aea', lat_0=35, lon_0=-80, llcrnrlon=-115, llcrnrlat=18, urcrnrlon=-63, urcrnrlat=55)
states = m.readshapefile('./tl_2010_us_state10', 'states')

# Latitudes and longitudes for "Hard to classify"
lats_hard = df_hard.Lat.values
lons_hard = df_hard.Long.values
x_hard, y_hard = m(lons_hard, lats_hard)

# Latitudes and longitudes for "New boil order issued or repaired"
lats_new = df_new.Lat.values
lons_new = df_new.Long.values
x_new, y_new = m(lons_new, lats_new)

# Latitudes and longitudes for "Drinking water contaminated"
lats_water = df_water.Lat.values
lons_water = df_water.Long.values
x_water, y_water = m(lons_water, lats_water)

# Latitudes and longitudes for "Existing boil order continuing or lifted"
lats_exist = df_exist.Lat.values
lons_exist = df_exist.Long.values
x_exist, y_exist = m(lons_exist, lats_exist)

# Latitudes and longitudes for "Beach/Lake/Park closed from bacteria in water"
lats_beach = df_beach.Lat.values
lons_beach = df_beach.Long.values
x_beach, y_beach = m(lons_beach, lats_beach)

# Adding color to the maps and markers
m.drawmapboundary(fill_color='#99ffff')
m.fillcontinents(color='#cc9966', lake_color='#99ffff')
m.plot(x_hard, y_hard, 'ro', markersize=8) # 'ro' = Red O's (circles) 
m.plot(x_new, y_new, 'go', markersize=8) # 'go' = Green O's (circles) 
m.plot(x_water, y_water, 'mo', markersize=8) # 'mo' = Red O's (circles) 
m.plot(x_exist, y_exist, 'bo', markersize=8) # 'bo' = Blue O's (circles) 
m.plot(x_beach, y_beach, 'yo', markersize=8) # 'yo' = Yellow O's (circles) 
```

**Beautiful.  All the yellow dots were actually beach closings around New Jersey and Long Island due to high bacteria counts in thw water!  Looks like we were able to change a single class data set into a multiclass data set!**
